```python
import numpy as np
import open3d as o3d
```

```python
dataname="pixeldata.xyz"
point_cloud= np.loadtxt(dataname,skiprows=1)
```

```python
pcd = o3d.geometry.PointCloud()
pcd.points = o3d.utility.Vector3dVector(point_cloud[:,:3])
pcd.colors = o3d.utility.Vector3dVector(point_cloud[:,3:6]/255)
pcd.normals = o3d.utility.Vector3dVector(point_cloud[:,6:9])
```

```python
import pyvista as pv
point_cloud1 = pv.PolyData(point_cloud)
point_cloud1
```

```python
point_cloud1.plot(eye_dome_lighting=True)
```

```python
point_cloud1.plot(render_points_as_spheres=True)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 10))
plt.scatter(point_cloud[:, 0], point_cloud[:, 1], c=point_cloud[:, 2])
plt.axis("image")
plt.xlabel("X Coordinate")
plt.ylabel("Y Coordinate")
#plt.show()
```

```python
import pix2vertex as p2v
from imageio import imread
image = imread('frame56.jpg')

result, crop = p2v.reconstruct(image)

p2v.vis_depth_interactive(result['Z_surface'])
```

```python
#p2v.vis_depth_matplotlib(crop,result['Z_surface'])
p2v.vis_depth_matplotlib(crop,result['Z_surface'],elevation=30,azimuth=45)
```

```python
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# load some test data for demonstration and plot a wireframe
X, Y, Z = axes3d.get_test_data(0.1)
ax.plot_wireframe(X, Y, Z, rstride=5, cstride=5)

# rotate the axes and update
for angle in range(0, 360):
    ax.view_init(30, angle)
    plt.draw()
    plt.pause(.001)
```
